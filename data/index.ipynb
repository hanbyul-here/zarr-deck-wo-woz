```python
import xarray as xr
import numpy as np
import rioxarray  # noqa

x = np.arange(-180, 180, 0.25)
y = np.arange(-90, 90.25, 0.25)
da = xr.DataArray(
    data=np.ones((len(y), len(x))),
    coords={"y": y, "x": x},
    dims=["y", "x"],
    name="my_array",
)
da = da * (np.abs(da.y) + np.abs(da.x) + 1)

ds = xr.Dataset({"my_array": da})
ds = ds.rio.write_crs("EPSG:4326")

ds.to_zarr('./test.zarr')

# pyramids = pyramid_reproject(ds, levels=6, resampling="bilinear")
# pyramids.to_zarr("./example.zarr", consolidated=True, mode="w")

```

```python
import numpy as np
import zarr

np.random.seed(0)


x = np.arange(-180, 180, 0.25)
y = np.arange(-90, 90.25, 0.25)

z = zarr.open(
    "dummy_dataset.zarr",
    shape=(len(y), len(x)),
    chunks=(1, 500),
    compressor=None,
    dtype=np.float32,
)

z[0] = np.arange(1000)
z[1] = np.sin(np.arange(1000) / 100)
z[2] = np.random.rand(1000)

print("Index 0:", z[0, :5])
print("Index 1:", z[1, :5])
print("Index 2:", z[2, :5])
```
